**Task 2** – Using the gene2go file from NCBI that we worked with in week 8 find the Gene Ontology terms that have been annotated to all of the SFARI genes (4marks).

```python
#retrieve the gene2go file
# date of last downloaded is 17th November 2020
ul.request.urlretrieve('https://ftp.ncbi.nlm.nih.gov/gene/DATA/gene2go.gz','gene2go1.gz')

gene2go = pd.read_csv('gene2go1.gz', compression='gzip', header=0, sep='\t')

gene2go.head()
```

```python
# test = gene2go[gene2go['GO_ID']  == 'GO:0032502']
# test
```

```python
annotated_genes = {'gene-symbol': [], 'gene-UID': [], 'go-ID': [], 'go-term': [], 'gene-score': []}
for gene_ID in gene_UID_df['gene-UID']:
    gene_sym = gene_UID_df.loc[gene_UID_df['gene-UID'] == gene_ID, 'gene-symbol'].iloc[0]
    gene_score = genes_df.loc[genes_df['gene-symbol'] == gene_sym, 'gene-score'].iloc[0]
    gene_annotation_df = gene2go[gene2go['GeneID']  == gene_ID]
    for go_ID in gene_annotation_df['GO_ID']:
        go_term = gene_annotation_df.loc[gene_annotation_df['GO_ID'] == go_ID, 'GO_term'].iloc[0]
        annotated_genes['gene-symbol'].append(gene_sym)
        annotated_genes['gene-score'].append(gene_score)
        annotated_genes['gene-UID'].append(gene_ID)
        annotated_genes['go-ID'].append(go_ID)
        annotated_genes['go-term'].append(go_term)
annotated_genes_df = pd.DataFrame.from_dict(annotated_genes)
    
```

**Task 1** – Create a plain text file of the NCBI UIDs for all of the gene-score 1 SFARI genes. Then, using the STRING website (https://string-db.org/) upload the gene-score 1 gene list, select “homo sapiens” as the species and hit search then click continue to visualise the protein-protein interaction network. In this network the genes are the nodes and the interactions between them the edges (connections). Click the “analysis” option near the bottom and report the following statistics: “number of nodes”, “number of edges” and “average node degree” (3 marks).

```python
from Bio import Entrez

Entrez.email = "Matthew.Whelan@ed.ac.uk"

# Storing the first Entrez ID number for each gene-symbol in a new df

gene_UID_dic = {'gene-symbol': [], 'gene-UID': []}
count = 0
genes_score_1_df = genes_df[genes_df['gene-score'] == 1.0]
genes_df = genes_score_1_df
for index, row in genes_df.iterrows():
#     print(row['gene-symbol'])

    # note the Entrez esearch function searches and returns a handle to the results.
    gene_symbol = row['gene-symbol']
    # restricting the search to gene, alive and human only
    handle = Entrez.esearch(db='gene',term=gene_symbol + '[gene] AND alive[prop] AND human[ORGN]',retmax=1,idtype="acc")
    record = Entrez.read(handle)
    gene_UID_dic['gene-symbol'].append(gene_symbol)
    try:
        gene_UID_dic['gene-UID'].append(record['IdList'][0])
    except:
        gene_UID_dic['gene-UID'].append(0)
    handle.close()
    count += 1
    progress = count / 993
#     if progress % 0.01 == 0:
    print(progress)

gene_UID_df = pd.DataFrame(gene_UID_dic)
gene_UID_df.head()

handle.close()
```

```python
gene_vals = gene_UID_df['gene-UID'].values
# np.savetxt('gene_UIDs.txt', gene_UID_df['gene-UID'].values)
gene_list = list(gene_vals)
for uid in gene_list:
    tfile = open('gene_UIDs.txt', 'a')
    tfile.write(uid + "\n")
    tfile.close()
```
